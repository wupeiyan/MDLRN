```python
import os
import numpy as np
from PIL import Image
import SimpleITK as sitk
```

```python
source_floder = 'data/origin/'
save_floder = 'data/nrrd/'
```

```python
for root, dirs, files in os.walk(source_floder, topdown=False):
    # for name in files:
    #     print(os.path.join(root, name))
    for name in dirs:
        if name == 'cdfi':
                continue
        os.makedirs(os.path.join(root, name).replace(source_floder, save_floder), exist_ok=True)

    for name in files:
        if 'cdfi' in root:
             continue
        file = os.path.join(root, name)
        
        print(file)
        print(file.replace(source_floder, save_floder).replace('png', 'nrrd'))
        img = np.asarray(Image.open(file).convert('L'))
        img = np.expand_dims(img, axis=2)
        img = sitk.GetImageFromArray(img) 
        sitk.WriteImage(img, file.replace(source_floder, save_floder).replace('png', 'nrrd'))
```
